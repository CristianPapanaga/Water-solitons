```python
from pydaqmx_helper.adc import ADC

# Initialise ADC class instance.
myADC = ADC()

# The output signal range is -4.5V to 4.5V (although do note the signal seems to often reach -5, 5 V), as stated in AWP-24-2
# manual. There are two wave height gauges which we  assign to channel 0 and 1, where 0 is the gauge closest to
# the wave generation point (the left-most side of the tank).
myADC.addChannels([0, 1],
               minRange = -4.5,
               maxRange = 4.5)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import scipy
import scipy.signal
from scipy import odr
import mpmath
import uncertainties as unc
from uncertainties import ufloat
from uncertainties import unumpy as unp

plt.rcParams.update({'font.size': 10})
plt.rcParams["figure.figsize"] = (6,4)
```

```python
##############################################################################################
## ----- PERFORM CALIBRATION AND DERIVE RELATIONSHIP BETWEEN VOLTAGE AND WATER HEIGHT ----- ##
##############################################################################################

# Calibration - a single point is taken at multiple known water heights in order to establish the relationship between 
# output voltage and water height.

#calibration_data = print(myADC.sampleVoltages(1, 100))

# Calibration heights in m, voltages in V.
# calibration_voltages0 corresponds to the left-most gauge, etc., and has been manually populated.
calibration_heights = np.array([0.03, 0.04, 0.05, 0.06, 0.07, 0.08, 0.09, 0.1])
calibration_voltages0 = [-4.925, -3.207, -2.017, 0.175, 1.164, 3.183, 4.195, 5.064]
calibration_voltages1 = [-4.885, -3.246, -2.118, -0.039, 1.071, 3.059, 4.032, 5.036]

# Define line for use in creating linear fit with scipy curve_fit().
def line(x, m, c):
    return m * x + c

x_fit = np.linspace(0.03, 0.1, 1000)

# Generate array of optimum line parameters popt, where popt[0] = m, popt[1] = c, and parameter covariance matrix pcov.
# For gauge 0.
popt0, pcov0 = scipy.optimize.curve_fit(line, calibration_heights, calibration_voltages0)

# Initialise correlated system of values for propagation handling by the Uncertainties package.
m0, c0 = unc.correlated_values(popt0, pcov0)
        
y_fit0 = line(x_fit, m0, c0)
nom_y0 = unp.nominal_values(y_fit0)
std_y0 = unp.std_devs(y_fit0)
        
# Plot linear fit line.
plt.plot(x_fit, nom_y0, color = "red", linewidth = 1)
        
# Plot 95% confidence interval, making sure to include covariance since fit parameters are obviously correlated.
lower_bound0 = nom_y0 - 2*std_y0
upper_bound0 = nom_y0 + 2*std_y0
plt.fill_between(x_fit, lower_bound0, upper_bound0, color = "black", alpha = 0.15, linewidth = 0)

# Scatter points for gauge 0.
plt.scatter(calibration_heights, calibration_voltages0, color = "black", zorder = 5)
plt.grid(linestyle = "--")
plt.title("Gauge 0 calibration")
plt.xlabel("Height (m)")
plt.ylabel("Output voltage (V)")
        
text0 = f"Linear fit parameters: \nm = {m0}\nc = {c0}"
plt.text(1.02, 0, text0, transform = ax.transAxes)
plt.legend(["Linear fit", "95% confidence band"])

    
plt.savefig("Gauge 0 Calibration Graph.pdf", bbox_inches = "tight")
plt.show()

# Generate array of optimum line parameters popt, where popt[0] = m, popt[1] = c, and parameter covariance matrix pcov.
# For gauge 1.
popt1, pcov1 = scipy.optimize.curve_fit(line, calibration_heights, calibration_voltages1)

# Initialise correlated system of values for propagation handling by the Uncertainties package.
m1, c1 = unc.correlated_values(popt1, pcov1)
        
y_fit1 = line(x_fit, m1, c1)
nom_y1 = unp.nominal_values(y_fit1)
std_y1 = unp.std_devs(y_fit1)
        
# Plot linear fit line.
plt.plot(x_fit, nom_y1, color = "red", linewidth = 1)
        
# Plot 95% confidence interval, making sure to include covariance since fit parameters are obviously correlated.
lower_bound1 = nom_y1 - 2*std_y1
upper_bound1 = nom_y1 + 2*std_y1
plt.fill_between(x_fit, lower_bound1, upper_bound1, color = "black", alpha = 0.15, linewidth = 0)

# Scatter points for gauge 1.
plt.scatter(calibration_heights, calibration_voltages1, color = "black", zorder = 5)
plt.grid(linestyle = "--")
plt.title("Gauge 1 calibration")
plt.xlabel("Height (m)")
plt.ylabel("Output voltage (V)")

text1 = f"Linear fit parameters: \nm = {m1}\nc = {c1}"
plt.text(1.02, 0, text1, transform = ax.transAxes)
plt.legend(["Linear fit", "95% confidence band"])

    
plt.savefig("Gauge 1 Calibration Graph.pdf", bbox_inches = "tight")
plt.show()

print("Height-voltage constant 0: ", m0)
print("Height-voltage constant 1: ", m1)
```

```python
###############################################################
## ----- TAKE MEASUREMENTS FROM THE WAVE HEIGHT GAUGES ----- ##
###############################################################

# Returns the water height from the measured voltage v, at the specified gauge, where gauge = 0 is the left-most gauge,
# gauge = 1 is the right-most.
def get_water_height(v, gauge):
    if (gauge == 0):
        return (v-c0)/m0
    elif (gauge == 1):
        return (v-c1)/m1

# Take samples from the wave height gauges, outputting heights from each gauge as a single array with two rows.
# How to use: run x, y = take_height_measurements() - gauge 0 values will be stored in x and gauge 1 values in y.
# n = number of total samples to take, r = samples to take per second. Samples are taken over n/r seconds (11s by default).
# Set save_data = True to save data to a .txt file - IMPORTANT: if doing this, specify also run_num to name
# the file appropriately (height_data_{run_num}.txt) and prevent previous runs being overwritten! Input also initial wave
# dimensions height A and length B, to save those. Also specify equilibrium water height eq_h.
def take_height_measurements(run_num, A, B, eq_h, n = 1100, r = 100, save_data = False):
    time_elapsed = n/r
    
    print(f"Samples being taken over {time_elapsed} seconds, please wait.")

    # Take n samples total of the voltage, taking r samples per second.
    raw_data = myADC.sampleVoltages(n, r)

    # Initialise an empty nparray for to transfer raw_data into for easier processing.
    data = np.empty([0])

    # Transfer data from standard array to numpy arrays.
    # data0 corresponds to the first (left-most) gauge while data1 corresponds to the right-most gauge.
    data0 = np.append(data, raw_data[0])
    data1 = np.append(data, raw_data[1])

    print(f"Samples succesfully taken, processing data.")
    
    ###################################
    ## ----- PROCESS WAVE DATA ----- ##
    ###################################

    # Initialise empty np arrays for storing height data.
    height_data0 = np.empty([0])
    height_data1 = np.empty([0])

    # Convert voltages to heights in m.
    for v in data0:
        h = get_water_height(v, 0)
        height_data0 = np.append(height_data0, h)
    
    for v in data1:
        h = get_water_height(v, 1)
        height_data1 = np.append(height_data1, h)
    
    if (save_data == True):
        header = f"Columns are height in m for gauge 0, gauge 1. \nInitial wave height A = {A} m, length B = {B} m."
        header += f"\nEquilibrium water height h = {eq_h} m. \n{n} total samples taken at {r} Hz over {time_elapsed} seconds."
        np.savetxt(
            f"height_data_{run_num}.txt",
            np.column_stack([height_data0, height_data1]),
            header = header
        )
        print(f"Samples successfully processed and saved to height_data_{run_num}.txt.")
    
        return np.array([height_data0, height_data1])
    
    else:
        print(f"Samples successfully processed, but not saved. Use save_data = True to save to .txt.")
    
        return np.array([height_data0, height_data1])

def normalize_height_data(run_num_start, run_num_end):
    for run_num in np.arange(run_num_start, run_num_end + 1):
        nom_height_data0, u_height_data0, nom_height_data1, u_height_data1 = np.loadtxt(f"height_data_{run_num}.txt",
                                                                                        unpack = "true")

        data_file = open(f"height_data_{run_num}.txt")
        for line in data_file:
            if "A" in line:
                initial_wave_params_line = line
            if "h" in line:
                water_equilibrium_height_line = line
            if "seconds" in line:
                time_elapsed_line = line

        # Look for any numbers in the words which constitute the water_equilibrium_height_line.
        # Once the single height value has been found, break from the loop as the other words don't need to be checked.
        for word in water_equilibrium_height_line.split():
            try:
                h = float(word)
                break
            except:
                pass

        # Split the initial_wave_params_line into individual words. Then, try and convert these words into float. If this is not
        # possible, pass. If it is, add the number to params.
        params = np.array([])
        for word in initial_wave_params_line.split():
            try:
                word = float(word)
                params = np.append(word, params)
            except:
                pass

        A = params[1]
        B = params[0]

        # Split the time_elapsed_line into individual words. Then, try and convert these words into float. If this is not
        # possible, pass. If it is, add the number to sampling_data.
        sampling_data = np.array([])
        for word in time_elapsed_line.split():
            try:
                word = float(word)
                sampling_data = np.append(word, sampling_data)
            except:
                pass

        n = sampling_data[2]
        r = sampling_data[1]
        time_elapsed = sampling_data[0]

        # Normalize height data such that equilibrium water height is always 0.
        norm_nom_height_data0 = nom_height_data0 - h
        norm_nom_height_data1 = nom_height_data1 - h

        header = f"Columns are height in m for gauge 0, gauge 1. \nInitial wave height A = {A} m, length B = {B} m."
        header += f"\nEquilibrium water height h = {h} m. \n{n} total samples taken at {r} Hz over {time_elapsed} seconds."

        np.savetxt(f"norm_height_data_{run_num}.txt",
                  np.column_stack([norm_nom_height_data0,
                                   u_height_data0,
                                   norm_nom_height_data1,
                                   u_height_data1]),
                  header = header)

# Access and plot the data stored in the file associated with run number run_num.
def open_and_plot_height_data(run_num, save_format = "png"):
    nom_height_data0, u_height_data0, nom_height_data1, u_height_data1 = np.loadtxt(f"norm_height_data_{run_num}.txt",
                                                                                    unpack = "true")
    
    # Get the time_elapsed from the data file.
    data_file = open(f"norm_height_data_{run_num}.txt")
    for line in data_file:
        if "seconds" in line:
            time_elapsed_line = line
            
    # Split the time_elapsed_line into individual words. Then, try and convert these words into float. If this is not
    # possible, pass. If it is, add the number to sampling_data.
    sampling_data = np.array([])
    for word in time_elapsed_line.split():
        try:
            word = float(word)
            sampling_data = np.append(word, sampling_data)
        except:
            pass
    
    time_elapsed = sampling_data[0]
    
    # Get the height A and length B of the initial wave from the data file.
    data_file = open(f"norm_height_data_{run_num}.txt")
    for line in data_file:
        if "A" in line:
            initial_wave_params_line = line
        elif "h" in line:
            water_equilibrium_height_line = line
            
    # Split the initial_wave_params_line into individual words. Then, try and convert these words into float. If this is not
    # possible, pass. If it is, add the number to params.
    params = np.array([])
    for word in initial_wave_params_line.split():
        try:
            word = float(word)
            params = np.append(word, params)
        except:
            pass
    
    A = params[1]
    B = params[0]
    
    a = A/2
    b = 2*B
    
    # By the same method as above, look for any numbers in the words which constitute the water_equilibrium_height_line.
    # Once the single height value has been found, break from the loop as the other words don't need to be checked.
    for word in water_equilibrium_height_line.split():
        try:
            h = float(word)
            break
        except:
            pass
    
    # Plot the first 5.5 seconds of data from gauge 0, then the last 5.5 seconds of data from gauge 1.
    plt.plot(np.linspace(0, time_elapsed/2, len(nom_height_data0[:len(nom_height_data0)//2])),
             nom_height_data0[:len(nom_height_data0)//2], color = "black", linewidth = 1)      
    plt.plot(np.linspace(time_elapsed/2, time_elapsed, len(nom_height_data1[len(nom_height_data1)//2:])),
             nom_height_data1[len(nom_height_data1)//2:], color = "black", linewidth = 1)
    plt.axvline(x = time_elapsed/2, linestyle = "--", color = "black")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude (m)")
    plt.grid(linestyle = "--")
    #plt.yticks([])
    
    ax = plt.gca()
    ax.invert_xaxis()
    
    tick_spacing = 1
    ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
    
    text1 = f"A = {A} m\nB = {B} m\nh = {h} m"
    text2 = "Gauge 0"
    text3 = "Gauge 1"
    plt.text(1.05, 0, text1, transform = ax.transAxes)
    plt.text(0.7, 1.05, text2, transform = ax.transAxes)
    plt.text(0.2, 1.05, text3, transform = ax.transAxes)
    
    # Add an arrow indicating positive x direction.
    ax.annotate("",
                xy=(1.02, -0.2),
                xycoords='axes fraction',
                xytext=(-0.02, -0.2),
                arrowprops=dict(arrowstyle="->", color='black'))
    plt.text(1.05, -0.213, "x", transform = ax.transAxes)
    
    plt.savefig(f"Run {run_num} Height Data Plot.{save_format}", bbox_inches = "tight")
    
    # Print header from data file.
    data_file = open(f"norm_height_data_{run_num}.txt")
    for line in data_file:
            if "#" in line:
                print(line)

################################################
## ----- ANALYSIS AND THEORETICAL STUFF ----- ##
################################################
        
# EQUATION 8: returns the amplitude of the wave given the max amplitude n_0,
# position x (ie. gauge positions and neighbourhood), wave speed c (calculated by wave_speed()), and the
# characteristic length of the wave L (calculated by char_length()).
def kdv_solitary_solution(n_0, x, t, c, L):
    return n_0 * (1/(unp.cosh((x-c*t)/L)))**2

# EQUATION 3: used in equation 9 below, where h is the equilibrium depth, g is acceleration due to gravity.
def get_c_0(g, h):
    return unp.sqrt(g*h)

# EQUATION 9: returns the wave speed given c_0 (calculated by c_0), equilibrium depth h, and max amplitude n_0.
def wave_speed(c_0, h, n_0):
    return c_0 * (1 + (n_0)/(2*h))

# EQUATION 10: returns the characteristic length L given equilibrium depth h, max amplitude n_0.
def char_length(h, n_0):
    return unp.sqrt(4*((h)**3)/(3*n_0))

# EQUATION 12: returns the time shifts dt1, dt2 caused by the collision of two solitons, 1 and 2, with speeds c_1, c_2
# in a water depth h.
def time_shift(c_1, c_2, n_0_1, n_0_2, h):
    K_1 = unp.sqrt((3*n_0_1)/(4*h))
    K_2 = unp.sqrt((3*n_0_2)/(4*h))
    dt1 = (h/c_1)*(1/K_1)*(unp.log((K_1+K_2)/(K_1-K_2)))
    dt2 = (h/c_2)*(1/K_2)*(unp.log((K_1+K_2)/(K_1-K_2)))
    return dt1, dt2

# EQUATION 11: returns the number of solitary waves which emerge from an initial rectangular wave of length b = 2B and
# height a = A/2, given the equilibrium depth h. Note that np.trunc(y) returns the integer part of y.
def number_of_solitons(a, h, b):
    S = unp.sqrt((3*a)/(2*h)) * b/h
    return 1 + unp.trunc(S/np.pi)

# Print the Ursell numbers for runs from run_num_start (inclusive) to run_num_end (exclusive).
def print_ursell_numbers(run_num_start, run_num_end):
    arr = np.array([])
    for i in range(run_num_start, run_num_end):
        # Get the height A and length B of the initial wave from the data file.
        data_file = open(f"norm_height_data_{i}.txt")
        for line in data_file:
            if "A" in line:
                initial_wave_params_line = line
            elif "h" in line:
                water_equilibrium_height_line = line

        # Look for any numbers in the words which constitute the water_equilibrium_height_line.
        # Once the single height value has been found, break from the loop as the other words don't need to be checked.
        for word in water_equilibrium_height_line.split():
            try:
                h = float(word)
                break
            except:
                pass

        # Split the initial_wave_params_line into individual words. Then, try and convert these words into float. If this is not
        # possible, pass. If it is, add the number to params.
        params = np.array([])
        for word in initial_wave_params_line.split():
            try:
                word = float(word)
                params = np.append(word, params)
            except:
                pass

        A = params[1]
        B = params[0]

        a = A/2
        b = 2*B

        # Calculate Ursell number.
        U = (b**2)*a/h**3

        N = get_number_of_solitons(run_num = i)

        print(A, B, N, U, "\n")

def get_measured_speed(run_num):
    nom_height_data0, u_height_data0, nom_height_data1, u_height_data1 = np.loadtxt(f"norm_height_data_{run_num}.txt",
                                                                                    unpack = "true")
    
    # Distance between the gauges in m, with an uncertainty for handling by the Uncertainties package.
    d_12 = ufloat(4.65, 0.05)
    
    # Find time taken to travel from gauge 0 to gauage 1. Note that since 100 samples are taken per second for 1100 total,
    # the time in seconds at, for example, peak_t_location_0 = 278 is t = 278/100 = 2.78 s.
    peak_t_locations_0, properties_0 = scipy.signal.find_peaks(nom_height_data0, height = [0.006, 0.06])
    peak_t_locations_1, properties_1 = scipy.signal.find_peaks(nom_height_data1, height = [0.006, 0.06])
    
    peak_heights_0 = np.sort(properties_0["peak_heights"])
    peak_heights_1 = np.sort(properties_1["peak_heights"])
    n_0 = peak_heights_0[-1]
    n_1 = peak_heights_1[-1]
    
    i_0 = 0
    for i in np.arange(0, len(nom_height_data0)):
        if (nom_height_data0[i] == n_0):
            i_0 = i
     
    i_1 = 0
    for i in np.arange(0, len(nom_height_data1)):
        if (nom_height_data1[i] == n_1):
            i_1 = i
            
    u_peak_heights = (u_height_data0[i_0], u_height_data1[i_1])
    
    time_0 = np.min(peak_t_locations_0)
    time_1 = np.min(peak_t_locations_1)
    
    time_taken = (time_1 - time_0)/100
    
    c_sp = d_12/time_taken
    
    peak_heights = (n_0, n_1)
    return c_sp, peak_heights, u_peak_heights

# Plot equation 9, for comparison with data. The argument 'runs' specifies the range of files to include in the plot. Set
# best_fit = True to generate a line of best fit.
def wave_speed_comparison(h = 0.06, runs = range(0, 38), best_fit = False, plot_errors = False):
    # Calculating c_0 with uncertainty on h.
    c_0 = get_c_0(9.8, ufloat(h, 0.005))

    measured_speeds = np.array([])
    avg_peak_heights = np.array([])

    for i in runs:
        c_sp, peak_height, u_peak_height = get_measured_speed(i)
        nom_n_0 = peak_height[0]
        u_n_0 = u_peak_height[0]
        nom_n_1 = peak_height[1]
        u_n_1 = u_peak_height[1]
        
        n_0 = ufloat(nom_n_0, u_n_0)
        n_1 = ufloat(nom_n_1, u_n_1)
        
        avg_n = (n_0 + n_1)/2

        avg_peak_heights = np.append(avg_n, avg_peak_heights)
        measured_speeds = np.append(c_sp, measured_speeds)

    #print("Average peak heights: ", avg_peak_heights)
    #print("Measured speeds: ", measured_speeds) 
    #print("Theoretical speeds: ", wave_speed(c_0, h, avg_peak_heights))
    
    x = unp.nominal_values(avg_peak_heights/h)
    xerr = unp.std_devs(avg_peak_heights/h)
    y = unp.nominal_values((measured_speeds/c_0)-1)
    yerr = unp.std_devs((measured_speeds/c_0)-1)
    
    # Plot best fit using an ODR method - this method is chosen due to the ability to generate parameters when there is
    # uncertainty on both the x and y axis variables.
    if(best_fit == True):
        data = odr.RealData(x, y, xerr, yerr)
        
        my_odr = odr.ODR(data, odr.unilinear)
        out = my_odr.run()
        
        # Output computed parameters, where popt[0] = m, popt[1] = c, covariance matrix pcov, and standard error perr.
        popt = out.beta
        pcov = out.cov_beta
        perr = out.sd_beta
        
        # Initialise correlated system of values for propagation handling by the Uncertainties package.
        m, c = unc.correlated_values(popt, pcov)
        
        x_fit = np.linspace(min(x), max(x), 1000)
        y_fit = line(x_fit, m, c)
        nom_y = unp.nominal_values(y_fit)
        
        # Plot linear fit line.
        plt.plot(x_fit, nom_y, color = "royalblue", linewidth = 1)
        
        # Plot 95% confidence band, making sure to include covariance since fit parameters are obviously correlated.
        # 95% confidence band parameters.
        popt_upper_bound = popt + 2 * perr
        popt_lower_bound = popt - 2 * perr
        
        fit_upper = line(x_fit, popt_upper_bound[0], popt_upper_bound[1])
        fit_lower = line(x_fit, popt_lower_bound[0], popt_lower_bound[1])
        plt.fill_between(x_fit, fit_lower, fit_upper, color = "black", alpha = 0.15, linewidth = 0)
        
        print(f"Linear fit line slope m = {m}.")

    # Plot theoretical line.
    plt.plot(x, unp.nominal_values(((wave_speed(c_0, h, avg_peak_heights))/c_0)-1), color = "red", linewidth = 1)
    
    print("Theoretical slope m = 0.5.")
    
    # Scatter data points.
    plt.scatter(x, y, c = "black", s = 10, zorder = 5)
    
    # Plot errorbars if prompted to.
    if (plot_errors == True):
        plt.errorbar(x, y, yerr, xerr, fmt = "none", ecolor = "gray", elinewidth = 0.6)
        
    text1 = f"Linear fit parameters:\nm = {m}\nc = {c}"
    text2 = f"Theoretical parameters:\nm = {0.5}\nc = {0}"
    plt.text(0.585, 0.075, text1)
    plt.text(0.585, 0, text2)
    plt.legend(["Linear fit", "95% confidence band", "Theoretical"])
    plt.xlabel(r"$\left\langle\eta_0\right\rangle\; / \;h$ (m)")
    plt.ylabel(r"$(c_{sp}\; / \;c_0) - 1$ (m/s)")
    plt.grid(linestyle = "--")
    
    plt.savefig("Wave speed comparison.pdf", bbox_inches = "tight")


# Gets the theoretically predicted number of solitary waves which should emerge from an initial rectangular wave of 
# length B and height A (in m), for comparison with real data.
def get_number_of_solitons(run_num):
    # Get the height A and length B of the initial wave from the data file.
    data_file = open(f"norm_height_data_{run_num}.txt")
    for line in data_file:
        if "A" in line:
            initial_wave_params_line = line
        elif "h" in line:
            water_equilibrium_height_line = line
            
    # Split the initial_wave_params_line into individual words. Then, try and convert these words into float. If this is not
    # possible, pass. If it is, add the number to params.
    params = np.array([])
    for word in initial_wave_params_line.split():
        try:
            word = float(word)
            params = np.append(word, params)
        except:
            pass
    
    A = params[1]
    B = params[0]
    
    a = A/2
    b = 2*B
    
    # By the same method as above, look for any numbers in the words which constitute the water_equilibrium_height_line.
    # Once the single height value has been found, break from the loop as the other words don't need to be checked.
    for word in water_equilibrium_height_line.split():
        try:
            h = float(word)
            break
        except:
            pass
    
    return number_of_solitons(a, h, b)
    
# Plot a KdV 1-soliton solution over a plot of height_data from run_num's gauge 0 or 1. The number of solitons predicted
# is determined and the correct number of theoretical solitons plotted. Note that no more than 2 solitons can
# currently be plotted. The sol_x_offset arguments can be used to account for velocity changes (and thus time taken changes)
# in the real waves due to energy loss.
def plot_theoretical_wave(run_num, t, t_1, gauge = None, sol_0_offset = 0, sol_1_offset = 0, plot_together = False):
    nom_height_data0, u_height_data0, nom_height_data1, u_height_data1 = np.loadtxt(f"norm_height_data_{run_num}.txt",
                                                                                    unpack = "true")
    
    # Get the equilibrium water height from the data file.
    data_file = open(f"norm_height_data_{run_num}.txt")
    for line in data_file:
        if "A" in line:
            initial_wave_params_line = line
        if "h" in line:
            water_equilibrium_height_line = line
        if "seconds" in line:
            time_elapsed_line = line
    
    # Look for any numbers in the words which constitute the water_equilibrium_height_line.
    # Once the single height value has been found, break from the loop as the other words don't need to be checked.
    for word in water_equilibrium_height_line.split():
        try:
            h = float(word)
            break
        except:
            pass
        
    # Split the initial_wave_params_line into individual words. Then, try and convert these words into float. If this is not
    # possible, pass. If it is, add the number to params.
    params = np.array([])
    for word in initial_wave_params_line.split():
        try:
            word = float(word)
            params = np.append(word, params)
        except:
            pass
    
    A = params[1]
    B = params[0]
            
    # Split the time_elapsed_line into individual words. Then, try and convert these words into float. If this is not
    # possible, pass. If it is, add the number to sampling_data.
    sampling_data = np.array([])
    for word in time_elapsed_line.split():
        try:
            word = float(word)
            sampling_data = np.append(word, sampling_data)
        except:
            pass
    
    time_elapsed = sampling_data[0]

    # Get highest two peaks from data, where n_0 is the highest - these should correspond to the soliton peaks.
    peak_times_g_0, properties_g_0 = scipy.signal.find_peaks(nom_height_data0, height = [-0.06, 0.06])
    peak_heights_g_0 = np.sort(properties_g_0["peak_heights"])
    n0_g_0 = np.abs(peak_heights_g_0[-1])
    n1_g_0 = np.abs(peak_heights_g_0[-2])

    peak_times_g_1, properties_g_1 = scipy.signal.find_peaks(nom_height_data1, height = [-0.06, 0.06])
    peak_heights_g_1 = np.sort(properties_g_1["peak_heights"])
    n0_g_1 = np.abs(peak_heights_g_1[-1])
    n1_g_1 = np.abs(peak_heights_g_1[-2])
    
    # Defining uncertainty on water depth measurement h.
    uh = ufloat(h, 0.005)
    
    # Calculating c_0 with uncertainty on h.
    c_0 = get_c_0(9.8, uh)
    
    # Calculate wave speed c of soliton with peak n0/n1.
    c0_g_0 = wave_speed(c_0, uh, n0_g_0)
    nom_c0_g_0 = unp.nominal_values(c0_g_0)
    
    c0_g_1 = wave_speed(c_0, uh, n0_g_1)
    nom_c0_g_1 = unp.nominal_values(c0_g_1)
    
    c1_g_0 = wave_speed(c_0, uh, n1_g_0)
    nom_c1_g_0 = unp.nominal_values(c1_g_0)
    
    
    c1_g_1 = wave_speed(c_0, uh, n1_g_1)
    nom_c1_g_1 = unp.nominal_values(c1_g_1)
    
    # Calculate characteristic wave length in x, L_x, and characteristic wave length in t, L_t, for two solitons.
    L_x0_g_0 = char_length(uh, n0_g_0)
    nom_L_x0_g_0 = unp.nominal_values(L_x0_g_0)
    
    L_x0_g_1 = char_length(uh, n0_g_1)
    nom_L_x0_g_1 = unp.nominal_values(L_x0_g_1)
    
    L_x1_g_0 = char_length(uh, n1_g_0)
    nom_L_x1_g_0 = unp.nominal_values(L_x1_g_0)
    
    L_x1_g_1 = char_length(uh, n1_g_1)
    nom_L_x1_g_1 = unp.nominal_values(L_x1_g_1)
    
    L_t0_g_0 = L_x0_g_0/c0_g_0
    nom_L_t0_g_0 = unp.nominal_values(L_t0_g_0)
    
    L_t0_g_1 = L_x0_g_1/c0_g_1
    nom_L_t0_g_1 = unp.nominal_values(L_t0_g_1)
    
    L_t1_g_0 = L_x1_g_0/c1_g_0
    nom_L_t1_g_0 = unp.nominal_values(L_t1_g_0)
    
    L_t1_g_1 = L_x0_g_1/c0_g_1
    nom_L_t1_g_1 = unp.nominal_values(L_t1_g_1)
    
    
    # Distances d_1, d_2 in m from the left-most edge to the gauges, and the times taken dt_1, dt_2.
    # Times taken allow for the wave to be plotted at the correct position on an amplitude-time graph
    # without manual correction or off-setting. Note, however, that since the waves will not behave like true solitons
    # due to energy loss to viscosity etc., an offset (in time taken) can be adjusted to account for such losses.
    d_1 = ufloat(1.945, 0.05)
    d_2 = ufloat(6.61, 0.05)
    dt_10_g_0 = d_1/c0_g_0 + sol_0_offset
    nom_dt_10_g_0 = unp.nominal_values(dt_10_g_0)
    
    dt_20_g_0 = d_2/c0_g_0 + sol_0_offset
    nom_dt_20_g_0 = unp.nominal_values(dt_20_g_0)
    
    dt_11_g_0 = d_1/c1_g_0 + sol_1_offset
    nom_dt_11_g_0 = unp.nominal_values(dt_11_g_0)
    
    dt_21_g_0 = d_2/c1_g_0 + sol_1_offset
    nom_dt_21_g_0 = unp.nominal_values(dt_21_g_0)
    
    
    dt_10_g_1 = d_1/c0_g_1 + sol_0_offset
    nom_dt_10_g_1 = unp.nominal_values(dt_10_g_1)
    
    dt_20_g_1 = d_2/c0_g_1 + sol_0_offset
    nom_dt_20_g_1 = unp.nominal_values(dt_20_g_1)
    
    dt_11_g_1 = d_1/c1_g_1 + sol_1_offset
    nom_dt_11_g_1 = unp.nominal_values(dt_11_g_1)
    
    dt_21_g_1 = d_2/c1_g_1 + sol_1_offset
    nom_dt_21_g_1 = unp.nominal_values(dt_21_g_1)
    
    
    # Get the number of solitons which should theoretically emerge from the initial disturbance. Note that only
    # two KdV solutions can be plotted at the moment.
    soliton_num = get_number_of_solitons(run_num)
    
    if (gauge == 0 and plot_together == False):
        if (soliton_num >= 2):
            plt.plot(t_1, kdv_solitary_solution(n1_g_0, nom_dt_11_g_0, t_1, nom_c1_g_0, nom_L_t1_g_0),
                     "--", color = "red", zorder = 5, alpha = 0.7)
            
        # Plot the KdV solution.
        plt.plot(t, kdv_solitary_solution(n0_g_0, nom_dt_10_g_0, t, nom_c0_g_0, nom_L_t0_g_0),
                 "--", color = "red", zorder = 5, alpha = 0.7)
        
        # Plot the height data for gauge 0.
        plt.plot(np.linspace(0, time_elapsed, np.size(nom_height_data0)), nom_height_data0, color = "black", linewidth = 1)
        
        plt.xlabel("Time (s)")
        plt.ylabel("Amplitude (m)")
        plt.grid(linestyle = "--")
        
        ax = plt.gca()
        ax.invert_xaxis()
        
        plt.text(1.05, 0, f"A = {A} m\nB = {B} m\nh = {h} m", transform = ax.transAxes)
        plt.text(0.06, 0.9, "Gauge 0", transform = ax.transAxes)
    
        # Add an arrow indicating positive x direction.
        ax.annotate("",
                    xy=(1.02, -0.2),
                    xycoords="axes fraction",
                    xytext=(-0.02, -0.2),
                    arrowprops=dict(arrowstyle="->", color="black"))
        plt.text(1.05, -0.213, "x", transform = ax.transAxes)
        
        tick_spacing = 1
        ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
        plt.savefig(f"Run {run_num} Gauge 0 Plot.pdf", bbox_inches = "tight")
        plt.show()

    elif (gauge == 1 and plot_together == False):
        if (soliton_num >= 2):
            plt.plot(t_1, kdv_solitary_solution(n1_g_1, nom_dt_21_g_1, t_1, nom_c1_g_1, nom_L_t1_g_1),
                     "--", color = "red", zorder = 5, alpha = 0.7)
            
        # Plot the KdV solution.
        plt.plot(t, kdv_solitary_solution(n0_g_1, nom_dt_20_g_1, t, nom_c0_g_1, nom_L_t0_g_1),
                 "--", color = "red", zorder = 5, alpha = 0.7)
        
        # Plot the height data from gauge 1.
        plt.plot(np.linspace(0, time_elapsed, np.size(nom_height_data1)), nom_height_data1, color = "black", linewidth = 1)
        
        plt.xlabel("Time (s)")
        plt.ylabel("Amplitude (m)")
        plt.grid(linestyle = "--")

        ax = plt.gca()
        ax.invert_xaxis()
        
        plt.text(1.05, 0, f"A = {A} m\nB = {B} m\nh = {h} m", transform = ax.transAxes)
        plt.text(0.83, 0.9, "Gauge 1", transform = ax.transAxes)
    
        # Add an arrow indicating positive x direction.
        ax.annotate("",
                    xy=(1.02, -0.2),
                    xycoords='axes fraction',
                    xytext=(-0.02, -0.2),
                    arrowprops=dict(arrowstyle="->", color='black'))
        plt.text(1.05, -0.213, "x", transform = ax.transAxes)
        
        tick_spacing = 1
        ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
        plt.savefig(f"Run {run_num} Gauge 1 Plot.pdf", bbox_inches = "tight")
        plt.show()
        
    elif (plot_together == True):
        # If there are more than two solitons, plot the second largest of those solitons for each gauge.
        if (soliton_num >= 2):
            plt.plot(t_1, kdv_solitary_solution(n1_g_0, nom_dt_11_g_0, t_1, nom_c1_g_0, nom_L_t1_g_0),
                     "--", color = "red", zorder = 5, alpha = 0.7)
            plt.plot(t_1, kdv_solitary_solution(n1_g_1, nom_dt_21_g_1, t_1, nom_c1_g_1, nom_L_t1_g_1),
                     "--", color = "red", zorder = 5, alpha = 0.7)
            
        # Plot KdV equation solution for gauge 0, and then gauge 1.
        plt.plot(t, kdv_solitary_solution(n0_g_0, nom_dt_10_g_0, t, nom_c0_g_0, nom_L_t0_g_0),
                 "--", color = "red", zorder = 5, alpha = 0.7)
        plt.plot(t, kdv_solitary_solution(n0_g_1, nom_dt_20_g_1, t, nom_c0_g_1, nom_L_t0_g_1),
                 "--", color = "red", zorder = 5, alpha = 0.7)
        
        # Plot the first half of the wave height data for the first gauge, then plot the last half of the wave height
        # data for the second gauge.
        plt.plot(np.linspace(0, time_elapsed/2, len(nom_height_data0[:len(nom_height_data0)//2])),
             nom_height_data0[:len(nom_height_data0)//2], color = "black", linewidth = 1)
        plt.plot(np.linspace(time_elapsed/2, time_elapsed, len(nom_height_data1[len(nom_height_data1)//2:])),
             nom_height_data1[len(nom_height_data1)//2:], color = "black", linewidth = 1)
        
        plt.axvline(x = time_elapsed/2, linestyle = "--", color = "black")
        plt.xlabel("Time (s)")
        plt.ylabel("Amplitude (m)")
        plt.grid(linestyle = "--")
        
        ax = plt.gca()
        ax.invert_xaxis()
        
        plt.text(1.05, 0, f"A = {A} m\nB = {B} m\nh = {h} m", transform = ax.transAxes)
        plt.text(0.7, 1.05, "Gauge 0", transform = ax.transAxes)
        plt.text(0.2, 1.05, "Gauge 1", transform = ax.transAxes)
    
        # Add an arrow indicating positive x direction.
        ax.annotate("",
                    xy=(1.02, -0.2),
                    xycoords='axes fraction',
                    xytext=(-0.02, -0.2),
                    arrowprops=dict(arrowstyle="->", color='black'))
        plt.text(1.05, -0.213, "x", transform = ax.transAxes)
        
        tick_spacing = 1
        ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
        plt.savefig(f"Run {run_num} Both Gauges Plot.pdf", bbox_inches = "tight")
        plt.show()
    
    # Print header from data file.
    data_file = open(f"norm_height_data_{run_num}.txt")
    for line in data_file:
        if "#" in line:
            print(line)
    
    #print(f"SOLITON 0 GAUGE 0 PARAMETERS: n0 = {n0_g_0}, c0 = {c0_g_0}, L_x0 = {L_x0_g_0}, L_t0 = {L_t0_g_0}.")
    #print(f"\nSOLITON 0 GAUGE 1 PARAMETERS: n0 = {n1_g_1}, c0 = {c0_g_1}, L_x0 = {L_x0_g_1}, L_t0 = {L_t0_g_1}.")
    #if (soliton_num >= 2):
    #    print(f"\nSOLITON 1 GAUGE 0 PARAMETERS: n1 = {n1_g_0}, c1 = {c1_g_0}, L_x1 = {L_x1_g_0}, L_t1 = {L_t1_g_0}.")
    #    print(f"\nSOLITON 1 GAUGE 1 PARAMETERS: n1 = {n1_g_1}, c1 = {c1_g_1}, L_x1 = {L_x1_g_1}, L_t1 = {L_t1_g_1}.")
        
    time_shifts = time_shift((c0_g_0+c0_g_1)/2, (c1_g_0+c1_g_1)/2, (n0_g_0+n0_g_1)/2, (n1_g_0+n1_g_1)/2, uh)
    print(f"Time shifts: {time_shifts}")
```
